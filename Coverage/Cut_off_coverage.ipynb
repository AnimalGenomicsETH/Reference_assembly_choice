```r
library(dplyr)
setwd("/cluster/work/pausch/temp_scratch/audald/variant_analyses/coverage")
UCD_summary = read.csv(file = "UCD/summary_coverage.txt", sep = "\t", header = TRUE)
UCD_cutoff = UCD_summary[(UCD_summary[,3]>9),] #Picking only samples with coverage in autosome over 9
UCD_ordered_cutoff = UCD_cutoff[with(UCD_cutoff, order(-autosome)),] #sorting the dataframe in a new environment (with) and descendent for the column (bam_id)
Angus_summary = read.csv(file = "Angus/summary_coverage.txt", sep = "\t", header = TRUE)
Angus_cutoff = Angus_summary[(Angus_summary[,3]>9),]
Angus_ordered_cutoff = Angus_cutoff[with(Angus_cutoff, order(-autosome)),]
print("Head of ordered samples as per coverage - UCD:")
head(UCD_ordered_cutoff)
print("Number of samples aligned to UCD over 9x:")
nrow(UCD_ordered_cutoff)
print("Head of ordered samples as per coverage - Angus:")
head(Angus_ordered_cutoff)
print("Number of samples aligned to Angus over 9x:")
nrow(Angus_ordered_cutoff)
UCD_ordered_cutoff_ID = UCD_ordered_cutoff[order(UCD_ordered_cutoff$bam_id),] #sorting the resulting rows by bam_id
Angus_ordered_cutoff_ID = Angus_ordered_cutoff[order(Angus_ordered_cutoff$bam_id),]
id_UCD = as.character(UCD_ordered_cutoff_ID$bam_id) #converting factors into characters for comparison
id_Angus = as.character(Angus_ordered_cutoff_ID$bam_id)
df <- data.frame(ucd = id_UCD,
                 angus = id_Angus) #merging the columns in the same dataframe for comparison
print("Are all the 161 samples included in both lists (UCD and Angus)?")
df$ucd == df$angus 
print("BAM_id of selected samples - UCD:")
print(id_UCD)
print("BAM_id of selected samples - Angus:")
print(id_Angus)
```
