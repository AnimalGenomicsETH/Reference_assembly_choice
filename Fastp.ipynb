## Observations and parameters for Fastp execution

This is how fastp is run with default parameters within Snakemake ([Snakefile](Snakemake/Snakefile.py), [Configuration details](Snakemake/config.yaml) and [Cluster details](Snakemake/cluster.json)):

```bash
/cluster/work/pausch/group_bin/fastp -i {input.R1} -o {output.R1_O} -I {input.R2} -O {output.R2_O} -h {output.R_HTML} -j {output.R_JSON} -q 15 -u 40 -g >/dev/null
```

The reports (JSON and HTML) show that not all the reads passed the quality control in all the cases. Even for those samples where all reads passed the quality control, some of the bases did not. This is probably due to a second application of the sanity check.

A summary of the information we can extract from the sanity check report by sample - script can be easily adapted to all samples:

```bash
module load jq/1.5
cd /cluster/work/pausch/temp_scratch/audald/best_assembly/fastp
samples="BSWCHEM110097189518 BSWCHEM110482036847 BSWCHEM120054800752"
for sample in $samples 
do  
    echo 'Filtering results for' ${sample}':'
    echo 'Total reads before filtering:'
    cat ${sample}/${sample}*json | jq '.summary.before_filtering'
    echo 'Total reads after filtering:'
    cat ${sample}/${sample}*json | jq '.summary.after_filtering'
    echo 'Filtering results:'
    cat ${sample}/${sample}*json | jq '.filtering_result'
    echo 'Duplication:'
    cat ${sample}/${sample}*json | jq '.duplication.rate'
    echo 'Insert size:'
    cat ${sample}/${sample}*json | jq '.insert_size.peak'
    echo ----------------------
    echo 
done
```
